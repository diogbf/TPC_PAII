```python
import csv
import numpy as np
import pandas as pd
pd.set_option('display.min_rows', 20)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cs = open('company_sales_data.csv')
CSdf = pd.read_csv(cs)

#CSdf = Company Sales data frame
#CSdf.head()

CSdf.head(12)

#Fiz head(12) para conseguir visualizar todos os dados uma vez que um ano só tem 12 meses.
```

<strong>Exercício 1:</strong>

```python
#Código para executar o gráfico:
CSdf.plot(kind="line", title= "Company profit per month", x="month_number", y="total_profit")

#Criação das listas:
profitList = CSdf ['total_profit'].tolist()
monthList = CSdf ['month_number'].tolist()

#Criação de nova escala:
plt.xticks(monthList)
plt.yticks(range(100000, 500002, 100000))

#Criação de legendas:
plt.ylabel('Profit in dollar')
plt.legend().remove()
```

<strong>Exercício 2:</strong>

```python
#Código para executar o gráfico segundo listas feitas no exercício 1:
x = monthList
y = profitList
plt.plot(x, y, color='r', linestyle='--', linewidth= 3, marker='o', markerfacecolor='k', label='Profit data of last year')

#Criação de nova escala:
plt.xticks(monthList)
plt.yticks(range(100000, 500002, 100000))


#legendas dos eixos X e Y e legenda do gráfico:
plt.xlabel('Número do mês')
plt.ylabel('Número de unidades vendidas')
plt.title("Company Sales data of last year")
plt.legend(loc='lower right')

plt.show()
```

<strong>Exercício 3:</strong>

```python
#Criação de novas variáveis:
mes = CSdf.month_number
face = CSdf.facecream
lavagem = CSdf.facewash
pasta_dentes = CSdf.toothpaste
sabonete = CSdf.bathingsoap
sh = CSdf.shampoo
hidratante = CSdf.moisturizer

#Redimensionar gráfico:
plt.figure(figsize=(8, 6))

#Definição de cada linha de dados:
plt.plot(mes, face, 'b', marker='o', label='Face cream sales data')
plt.plot(mes, lavagem, 'orange', marker='x', label='Face wash sales data')
plt.plot(mes, pasta_dentes, 'g', marker='o', label='Toothpaste sales data')
plt.plot(mes, sabonete, 'r', marker='o', label='Bathing soap sales data')
plt.plot(mes, sh, 'purple', marker='o', label='Shampoo sales data')
plt.plot(mes, hidratante, 'c', marker='o', label='Moisturizer soap sales data')

#Criação de nova escala:
plt.xticks(monthList)

#legendas:
plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('Sales data', size=(18))
plt.legend(loc='upper left', prop={'size': 11})

plt.show()
```

<strong>Exercício 4:</strong>

```python
#Novas variáveis, soma do valor mensal:
face_s = CSdf.facecream.sum()
lavagem_s = CSdf.facewash.sum()
pasta_dentes_s = CSdf.toothpaste.sum()
sabonete_s = CSdf.bathingsoap.sum()
sh_s = CSdf.shampoo.sum()
hidratante_s = CSdf.moisturizer.sum()

#Criação de listas:
lista_vendas = [face_s, lavagem_s, pasta_dentes_s, sabonete_s, sh_s, hidratante_s]
legendas = ['face cream', 'face wash', 'toothpaste', 'bathing soap', 'shampoo', 'moisturizer']

#Para formar o gráfico com percentagens:
plt.pie(lista_vendas, labels = legendas, autopct='%1.1f%%')

#Legendas(optei por uma legenda diferente dos produtos para o gráfico não perder visibilidade):
plt.title("Sales data", size=18)
plt.legend(legendas, title='Produtos', loc="center left", bbox_to_anchor=(1.25, 0, 1, 1))

#Legenda para ficar igual ao enunciado:
#plt.legend(legendas, loc='lower right')

plt.show()
```
