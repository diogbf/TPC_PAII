```python
import csv
import numpy as np
import pandas as pd
pd.set_option('display.min_rows', 20)
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
cs = open('company_sales_data.csv')
companyDF = pd.read_csv(cs)
companyDF.head(12)
```

```python
#Exercicio 1

companyDF.plot(kind="line", x="month_number", y= "total_profit")
plt.xticks(range(1,13,1))
plt.yticks(range(100000,500001,100000))
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.title('Company profit per month')
plt.legend().remove()
plt.show()

profitList = companyDF['total_profit'].tolist()
monthList = companyDF['month_number'].tolist()
```

```python
#Exercicio 2

plt.plot(monthList, profitList, 'r--', label="Profit data of last year", linewidth=3.0, marker='o',markerfacecolor='k',markeredgecolor='k')
plt.xlabel('Nmero do ms')
plt.ylabel('Nmero de unidades vendidas')
plt.title('Company Sales data of last year ')
plt.xticks(monthList)
plt.yticks(range(100000,500001,100000))
plt.legend(loc='lower right')
plt.show()
```

```python
#Exercicio 3
month = companyDF.month_number 
fc = companyDF.facecream
fw = companyDF.facewash
t = companyDF.toothpaste
b = companyDF.bathingsoap
sh = companyDF.shampoo
m = companyDF.moisturizer

plt.plot(month, fc, 'b',marker='o',label= "Face cream Sales Data")
plt.plot(month, fw, 'y',marker='o',label= "Face wash Sales Data")
plt.plot(month, t, 'g',marker='o',label= "ToothPaste Sales Data")
plt.plot(month, b, 'r',marker='o',label= "ToothPaste Sales Data")
plt.plot(month, sh, 'c',marker='o',label= "ToothPaste Sales Data")
plt.plot(month, m, 'm',marker='o',label= "ToothPaste Sales Data")
plt.xticks(range(1,13,1))
plt.yticks(range(2000,18001,2000))
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.title('Sales data')
plt.legend(loc='upper left')
plt.show()
```

```python
#Exercicio 4
fc_sum = fc.sum()
fw_sum = fw.sum()
t_sum = t.sum()
b_sum = b.sum()
sh_sum = sh.sum()
m_sum = m.sum()
somas = [fc_sum,fw_sum,t_sum,b_sum,sh_sum,m_sum]
my_labels = 'Facecream','Facewash','Toothpaste','Bathingsoap','Shampoo','Moisturizer'
plt.pie(somas,labels=my_labels,autopct='%1.1f%%')
plt.title ('Sales data')
plt.legend(loc='lower right')
plt.show ()
```
